```python
import numpy as np
import pandas as pd
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf 
from statsmodels.tsa.seasonal import seasonal_decompose 
from pmdarima import auto_arima                        
from sklearn.metrics import mean_squared_error
from statsmodels.tools.eval_measures import rmse

from statsmodels.tsa.arima_model import ARIMA


import warnings

from matplotlib import pyplot

# from pyramid.arima import auto_arima
from pandas.plotting import autocorrelation_plot
from math import sqrt

```

```python
def parsealldate(x):
    timestamp = pd.datetime.strptime(x,'%Y-%m-%d %H:%M:%S')
#     timestamp = timestamp.dt.ceil(freq='s') 
    return timestamp
```

```python
def runArima(trainData, testData, pdq, modelName, solver):
    
    model = ARIMA(trainData, order=pdq)
    model_fit = model.fit(solver = solver)
    model_fit.save('./arimaModels/'+ modelName + '.pkl')

    history = [x for x in trainData]
    
    predictions = list()
# walk-forward validation
    for t in range(len(testData)):

        walkForwardModel = ARIMA(history, order = pdq)
        walkForwardModel_fit = walkForwardModel.fit(solver = solver,disp=False)
        output = walkForwardModel_fit.forecast()
        yhat = output[0]
        predictions.append(yhat)
        obs = testData[t]
        history.append(obs)
        

    return predictions
```

```python
sampleRate = 10
serverFile = 1
trainFileName = "../parse_tactics_weekend_sample_rate/normalized_tva_server_"+ str(serverFile) + "_tactic_1_train_weekdayOneHot.csv"
testFileName = "../parse_tactics_weekend_sample_rate/normalized_tva_server_"+ str(serverFile) + "_tactic_1_test_weekdayOneHot.csv"
validationFileName = testFileName = "../parse_tactics_weekend_sample_rate/normalized_tva_server_"+ str(serverFile) + "_tactic_1_validation_weekdayOneHot.csv"
```

```python
trainData = pd.read_csv(trainFileName,  parse_dates=[0])
validationData = pd.read_csv(validationFileName, parse_dates=[0])
trainData['timestamp'] = trainData.timestamp.dt.ceil(freq='s')
validationData['timestamp'] = validationData.timestamp.dt.ceil(freq='s')
```

```python
trainData = trainData.set_index('timestamp')['cost']
validationData = validationData.set_index('timestamp')['cost']
trainData = trainData.dropna()
```

```python

# bestModel = auto_arima(trainData, trace=True, error_action='ignore', suppress_warnings=True,method='nm')
bestModel = (1,1,0)
testFileName = "../parse_tactics/normalized_tva_server_"+ str(serverFile) +"_tactic_1_test.csv"

testData = pd.read_csv(testFileName, parse_dates=[0])
testData['timestamp'] = testData.timestamp.dt.ceil(freq='s')
testData = testData.set_index('timestamp')['cost']
testData.index = pd.DatetimeIndex(testData.index).to_period('s')


predictions = runArima(trainData, testData, bestModel,'cost_server_' + str(serverFile), solver = 'nm')

```

```python
sampleRate = 10
predictions = []

for server in range(0,3):

    trainFileName = "../parse_tactics_weekend_sample_rate/normalized_tva_server_"+ str(server+1) + "_tactic_1_train_weekdayOneHot.csv"
    testFileName = "../parse_tactics_weekend_sample_rate/normalized_tva_server_"+ str(server+1) + "_tactic_1_test_weekdayOneHot.csv"
    validationFileName = testFileName = "../parse_tactics_weekend_sample_rate/normalized_tva_server_"+ str(server+1) + "_tactic_1_validation_weekdayOneHot.csv"
    
    trainData = pd.read_csv(trainFileName,  parse_dates=[0])
    validationData = pd.read_csv(validationFileName, parse_dates=[0])
    trainData['timestamp'] = trainData.timestamp.dt.ceil(freq='s')
    validationData['timestamp'] = validationData.timestamp.dt.ceil(freq='s')

    trainData = trainData.set_index('timestamp')['cost']
    validationData = validationData.set_index('timestamp')['cost']
    trainData = trainData.dropna()
    

#     bestModel = auto_arima(trainData, trace=True, error_action='ignore', suppress_warnings=True,method='nm')
    bestModel = (1,1,0)
    
    testData = pd.read_csv(testFileName, parse_dates=[0])
    testData['timestamp'] = testData.timestamp.dt.ceil(freq='s')
    testData = testData.set_index('timestamp')['cost']
    testData.index = pd.DatetimeIndex(testData.index).to_period('s')
    
    
    
    
    predictionFileName = './predictions/predictions_server_' + str(server+1) + '_tactic_1_cost_sampleRate_'+ str(sampleRate) + '.csv'
    modelFileName = 'cost_server_' + str(server+1) + '_sample_' + str(sampleRate)
    

    prediction = runArima(trainData, testData, bestModel, modelFileName, solver = 'nm')
    pd.DataFrame(np.array(prediction).flatten(), columns=['predicted_Cost']).to_csv(predictionFileName, index = False)    
    
    predictions.append(prediction)
```
