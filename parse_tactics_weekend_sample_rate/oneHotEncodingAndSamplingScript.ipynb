```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
```

## OneHotEncoding DayOfWeek

```python
def addDayOfWeekToData(data):
    encoder = OneHotEncoder(sparse=False)
    encoder.fit(np.array(range(0,7)).reshape(-1, 1))
    dayOfWeek = pd.DataFrame(encoder.transform((np.array(data.timestamp.dt.dayofweek).reshape(-1,1))), columns= ['day'+str(i+1) for i in range(0, 7)])
    return pd.concat([data, dayOfWeek], axis=1)
```

```python
for server in range(1,4):
    for fType in ['train', 'test', 'validation']:
        fileName = "normalized_tva_server_" + str(server) + "_tactic_1_" + fType
        data = pd.read_csv( "../parse_tactics/" + fileName + ".csv", parse_dates=[0])
        data.head()
        print("../parse_tactics/" + fileName + ".csv")
        data = addDayOfWeekToData(data)
        data.to_csv("./" + fileName + "_weekdayOneHot.csv",index = False)
```

## Sampling rate

```python
def sampleData(data, sampleRate):
    newData = []
    for i in range(0,len(data),sampleRate):
        newData.append(data.iloc[i])
    print(len(newData))
    newData = pd.DataFrame(newData, columns=data.columns)
    return newData
```

```python
for server in range(1,4):
    for fType in ['train', 'test', 'validation']:
        for rate in [5,10,20]:
            
            fileName = "normalized_tva_server_" + str(server) + "_tactic_1_" + fType + "_weekdayOneHot"
            data = pd.read_csv( "./" + fileName + ".csv", parse_dates=[0])


#             print(len(data))
            data = sampleData(data, rate)

            finalFileName = "./sampleRate/" + fileName + "sample_rate_" + str(rate) + ".csv"
#             print(finalFileName)

            data.to_csv(finalFileName, index = False)
        
        
```
