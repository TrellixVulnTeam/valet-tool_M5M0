```python

def combineExamResults(path,serverNumber,pingOrNo = False):
    noPingvalidationFile = 'noping_normalized_tva_server_'+str(serverNumber)+'_tactic_1_validation_predictions.csv'
    pingvalidationFile = 'ping_normalized_tva_server_'+str(serverNumber)+'_tactic_1_validation_predictions.csv'
    
    validationFile = noPingvalidationFile
    fileBaseName = './eRnnFinalFiles/noping_normalized_server_'
    if pingOrNo == True:
        validationFile = pingvalidationFile
        fileBaseName = './eRnnFinalFiles/ping_normalized_server_'
    
    costPath = path+'tva_server_'+str(serverNumber)+'_cost/'
    latencyPath = path+'tva_server_'+str(serverNumber)+'_latency/'
    
    costFile = (pd.read_csv(costPath+validationFile))
    latencyFile = (pd.read_csv(latencyPath+validationFile))
    
    costFile['expected_latency'] = latencyFile['expected_latency'] 
    costFile['predicted_latency'] = latencyFile['predicted_latency'] 
    
    
    parseTacticFiles = '../parse_tactics/normalized_tva_server_'+str(serverNumber)+'_tactic_1_validation.csv'
    rawData = pd.read_csv(parseTacticFiles)
    timestamp = pd.to_datetime(rawData.timestamp.iloc[1:])
    
    costFile['Hours'] = timestamp.dt.hour
    costFile['Minutes'] = timestamp.dt.minute
    costFile['Seconds'] = timestamp.dt.second


    
    
    
    costFile.to_csv(fileBaseName+str(serverNumber)+'_tactic_1_validation.csv',index=False)
```

```python
for i in range(1,4):
    combineExamResults(predictionInputPath,i)
```
