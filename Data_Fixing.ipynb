```python
import pandas as pd
import numpy as np
from sklearn import preprocessing

data = pd.read_csv("document1.csv")

data['Timestamp'] = pd.to_datetime(data['Timestamp'])
```

```python
data.head()
```

```python
df=data.loc[(data['Tactic']==1) | (data['Tactic']==5)]
```

```python
i=1
j=2
while(i<df.shape[0] and j<df.shape[0]):
    
    check=df.iloc[j].Timestamp-df.iloc[i].Timestamp
    check = check.total_seconds()
    if(df.iloc[j,6] != -1):
        df.iloc[j,4] = check
    i+=2
    j+=2

```

```python
data.set_index(['Timestamp'], inplace=True)
data.update(df.set_index(['Timestamp']))
df.reset_index(inplace=True)
```

```python
data.drop(['Unnamed: 0'],inplace = True,axis =1)
data.drop(['Unnamed: 7'],inplace = True,axis =1)
```
