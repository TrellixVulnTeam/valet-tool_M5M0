```python
for server in range(1,4):
        
    nRate = -1

    for rate in [0, 0, 5, 10, 20]:

        arimaData = pd.read_csv('./final/arima/' + str(server) + '/arima_predictions_server_' + str(server) + '_tactic_1_sampleRate_' + str(rate) + '.csv')

        timeSeries = arimaData['timestamp'].iloc[:-1]
        realCost = arimaData['cost'].iloc[:-1]
        realLatency = arimaData['latency'].iloc[:-1]
        realReliability = arimaData['reliability'].iloc[:-1]


        svrRbfChangeFile = '../niranjana/results/' + str(server) + '/raw_results_svr_rbf_iteration_'  + str(nRate) + '.npy'
    
        svrLinearChangeFile = '../niranjana/results/' + str(server) + '/raw_results_svr_linear_iterations'  + str(nRate) + '.npy'
        
        
#         print(toChangeFile)


        rbfData = np.load(svrRbfChangeFile)
        rbfData = pd.DataFrame(rbfData, columns=['predicted_latency', 'predicted_cost', 'predicted_reliability'])
        
        linearData = np.load(svrLinearChangeFile)
        linearData = pd.DataFrame(linearData, columns=['predicted_latency', 'predicted_cost', 'predicted_reliability'])

        
        rbfData['timestamp'] = timeSeries
        rbfData['cost'] = realCost
        rbfData['latency'] = realLatency
        rbfData['reliability'] = realReliability
        
        linearData['timestamp'] = timeSeries
        linearData['cost'] = realCost
        linearData['latency'] = realLatency
        linearData['reliability'] = realReliability

        print(rbfData.shape)
        
        
        if nRate == -1:
            nRate += 2
            rbfSaveFile = './final/svr_rbf/' + str(server) + '/svr_rbf_predictions_server_' + str(server) + '_tactic_1_sampleRate_' + str(rate) + '_no_ping.csv'
            linearSaveFile = './final/svr_linear/' + str(server) + '/svr_linear_predictions_server_' + str(server) + '_tactic_1_sampleRate_' + str(rate) + '_no_ping.csv'
        
        else:
            nRate += 1
            rbfSaveFile = './final/svr_rbf/' + str(server) + '/svr_rbf_predictions_server_' + str(server) + '_tactic_1_sampleRate_' + str(rate) + '.csv'
            linearSaveFile = './final/svr_linear/' + str(server) + '/svr_linear_predictions_server_' + str(server) + '_tactic_1_sampleRate_' + str(rate) + '.csv'
        


        
        
        rbfData.to_csv(rbfSaveFile, index = False)
        linearData.to_csv(linearSaveFile, index = False)


        
```

```python
#renaming col headers
for (dirpath, dirnames, filenames) in os.walk('./final/lstm/'):
    
    for file in filenames:
        if '.csv' in file:
            data = pd.read_csv(os.path.join(dirpath, file))
            print(file)
            renameCols = {}
            for col in data.columns:

                if 'cost' in col.lower():

                    if 'predicted' in col.lower():
#                         data['predicted_cost'] = data[col]
                        renameCols[col] = 'predicted_cost'
                    else:
#                         data['cost'] = data[col]
                        renameCols[col] = 'cost'

                elif 'latency' in col.lower():

                    if 'predicted' in col.lower():
#                         data['predicted_latency'] = data[col]
                        renameCols[col] = 'predicted_latency'

                    else:
#                         data['latency'] = data[col]
                        renameCols[col] = 'latency'
                elif 'reliability' in col.lower():

                    if 'predicted' in col.lower():
#                         data['predicted_latency'] = data[col]
                        renameCols[col] = 'predicted_reliability'

                    else:
#                         data['latency'] = data[col]
                        renameCols[col] = 'reliability'
#             print(renameCols)
            data.rename(columns=renameCols,inplace=True)
    
            data = data.iloc[:-1]
            data.to_csv(os.path.join(dirpath, file), index = False)
```
