```python
# os.walk('./celia/')
for (dirpath, dirnames, filenames) in os.walk('./final/mlp/'):
    
    for file in filenames:
        if '.csv' in file:
            data = pd.read_csv(os.path.join(dirpath, file))
            print(file)
            renameCols = {}
            for col in data.columns:

                if 'cost' in col.lower():

                    if 'predicted' in col.lower():
#                         data['predicted_cost'] = data[col]
                        renameCols[col] = 'predicted_cost'
                    else:
#                         data['cost'] = data[col]
                        renameCols[col] = 'cost'

                elif 'latency' in col.lower():

                    if 'predicted' in col.lower():
#                         data['predicted_latency'] = data[col]
                        renameCols[col] = 'predicted_latency'

                    else:
#                         data['latency'] = data[col]
                        renameCols[col] = 'latency'
                elif 'reliability' in col.lower():

                    if 'predicted' in col.lower():
#                         data['predicted_latency'] = data[col]
                        renameCols[col] = 'predicted_reliability'

                    else:
#                         data['latency'] = data[col]
                        renameCols[col] = 'reliability'
#             print(renameCols)
            data.rename(columns=renameCols,inplace=True)
    
            data = data.iloc[:-1]
            data.to_csv(os.path.join(dirpath, file), index = False)
```
